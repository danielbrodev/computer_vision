````python
import matplotlib
print(matplotlib.__version_info__)
````

````python
import matplotlib.pyplot as plt # pyplot bu matplotlib ning sub modeli bolib biz chizmoqchi olgan graphlarni chizishga yordam beradi
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5.6, 6, 8, 0])

plt.plot(x, y) #
plt.title("Basic Line Plot example")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.show()
````

````python
from matplotlib.lines import lineStyles
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5.6, 6, 8, 0])

plt.plot(y, linestyle='dotted', color='green', marker='o') # bu yerda x argument berilmayabdi graph uni qayerdan olyabdi???
plt.title("Styled Line Plot example")
plt.xlabel("X-value")
plt.ylabel("Y-value")
````

````python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 5.6, 6, 8, 0])
y2 = np.array([3, 6.6, 9, 0, -1])

plt.plot(x, y1, label ='Line 1', color='red', marker='s')
plt.plot(x, y2, label ='Line 2', color='green', marker='o')
plt.title("Styled Line Plot example")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.legend() # bu bizga graphlarni ajratib beruvchi mark
plt.show()
````

````python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.title("Bar Chart example")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.show()
````

````python
x = np.array([5, 7, 8, 7])
y = np.array([99, 86, 87, 89])

plt.scatter(x, y, color='purple')
plt.title("Scatter Plot example")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.show()
````

# **Histogram with Numpy and Mathlotlib**

````python
data = np.array([3, 4, 5, 6, 7])
plt.hist(data, bins=8, color='red')
plt.title("Histogram example")
plt.xlabel("X-value")
plt.ylabel("Y-value")
plt.show()
````

# **Subplots(Multiple Graphs Together)**

````python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 5.6, 6, 8, 0])
y2 = np.array([3, 6.6, 9, 0, -1])

plt.subplot(1, 2, 1)
plt.plot(x, y1, label ='Line 1', color='red', marker='s')
plt.title("Styled Line Plot example")
````

````python
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 5.6, 6, 8, 0.2])
y2 = np.array([3, 6.6, 9, 0.5, 1])
plt.figure(figsize=(10, 5)) # with height

plt.subplot(1, 2, 1)
plt.plot(x, y1, label ='Line 1', color='red', marker='s')
plt.title("Line 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, label= "Line 2")
plt.title("Line 2")

plt.tight_layout()
plt.show()

````

````python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 5.6, 6, 8, 0.2])

plt.plot(x, y, color='green')
plt.title('Save Figure Example')
plt.savefig('my_figure.png')
plt.show()
````
