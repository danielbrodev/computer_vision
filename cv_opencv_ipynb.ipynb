```python
from google.colab.patches import cv2_imshow
import cv2
image = cv2.imread('cat.jpeg')
cv2_imshow(image)
```

```python
print("shape:", image.shape) # height , width, color channels (Red Green Blue)

print("H:", image.shape[0])
print("W:", image.shape[1])
print("C:", image.shape[2])
```

```python
(b, g, r)= image[100, 50] # bu 100 - qatordagi 50 - ustundagi ranglarni olib beradi
print("Blue", b, "Green", g, "Red", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resize = cv2.resize(image, (200, 200))
cv2_imshow(resize)
# rasm olchamini o'zgartira olamiz
```

```python
cropped = image[100:300, 150:400]
cv2_imshow(cropped)
```

```python
roteted = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(roteted)
```

```python
flipped = cv2.flip(image, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("newimage.jpg", flipped)
```

```python
cropped_cat = cv2.imread("newimage.jpg")
cv2_imshow(cropped_cat)
```

```python
print("Hello World")
 # pyrefly: ignore [missing-import]
import cv2
cap = cv2.VideoCapture(0) # 0 bu yerda default birinchi kamerani ishga tushir

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

while True:
    success, frame = cap.read()
    if not success:
        print("Failed to capture frame")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.5, 4)
    for x,y,w,h in faces:
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0,255,0), 2)

    cv2.imshow("Real time", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()










# ==============================

# # pyrefly: ignore [missing-import]
# import cv2
# cap = cv2.VideoCapture(0)


# while True:
#     success, frame = cap.read()

#     if not success:
#         print("Failed to capture frame")
#         break

#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
#     blurred = cv2.GaussianBlur(frame, (21, 21), 0)
#     edges1 = cv2.Canny(frame, 100, 200)
#     edges2 = cv2.Canny(frame, 50, 250)

#     flipped = cv2.flip(frame, 0)
#     # cv2.imshow("Frame", frame)
#     # cv2.imshow("Gray", gray)
#     cv2.imshow("Flipped", flipped)


#     if cv2.waitKey(1) & 0xFF == ord("q"):
#         break

# cap.release()
# cv2.destroyAllWindows()

# ==============================

```
